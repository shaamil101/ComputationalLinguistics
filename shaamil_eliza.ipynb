The program uses regular expressions to parse a user's input and generate a  response in Spanish


```python
# Import libraries for string handling (string) and regular expressions (re)
import string
import re
```

```python
# =====================================================================================
# Function:     eliza
#
# Input:        userInput (a string with a phrase in the English language)
# Output:       reply (a string with a response, in Spanish, generated by the function)
# Description:  The function uses regular expressions to read input from the user
#               and produce a response using natural language in Spanish. We are able to account for 8 different type of conversation sentences for now.
# =====================================================================================

def eliza(userInput):

    # Create the variable output, which will contain our response
    output = ""

    # Create the regular expressions
    # You can test your regular expressions in numerous websites, such as https://regex101.com/

    reHello1 = r"(?:Mi nombre es|Me llamo) (.*)[.!]"


    # capturing either no estoy or estoy if present
    reHello2 = r"(no estoy|estoy).*\b(\w+)\b"


    #Assuming that the sentences always ends with period or exclamation
    reHello3 = r"(no soy|soy) (.*)[.!]"

    reHello4 = r"(mamá|papá)"
    reHello5 = r"(no quiero|quiero|no debo|debo|no puedo|puedo) (.*)[.!]"
    reHello6 = r"(pienso|espero)"
    reHello7 = r"\bsiempre\b"
    reHello8 = r"(estúpida|idiota)"



    groupHello1 = re.search(reHello1,userInput,re.IGNORECASE)
    groupHello2 = re.search(reHello2,userInput,re.IGNORECASE)
    groupHello3 = re.search(reHello3,userInput,re.IGNORECASE)
    groupHello4 = re.search(reHello4,userInput,re.IGNORECASE)
    groupHello5 = re.search(reHello5,userInput,re.IGNORECASE)
    groupHello6 = re.search(reHello6,userInput,re.IGNORECASE)
    groupHello7 = re.search(reHello7,userInput,re.IGNORECASE)
    groupHello8 = re.search(reHello8,userInput,re.IGNORECASE)

    # if the input matches the regular expression:
    #checking for name
    if (groupHello1 != None):
        output = "Hola, " + groupHello1.group(1) + ". Como estas?"
    #checking for no estoy or estoy
    elif (groupHello2 != None):
        if (groupHello2.group(1) == "no estoy"):
            output = "Porque no estas " + groupHello2.group(2) + "?"
        else:
            output = "Porque estas " + groupHello2.group(2) + "?"
    #checking for no soy or soy
    elif (groupHello3 != None):
        if (groupHello3.group(1) == "no soy"):
            output = "Porque no eres " + groupHello3.group(2) + "?"
        else:
            output = "Porque eres " + groupHello3.group(2) + "?"
    #checking for mama or papa
    elif (groupHello4 != None):
        if (groupHello4.group(1).lower() == "mamá"):
            output = "Cuentame más de tu mamá."
        else:
            output = "Cuentame más de tu papá."
    #checking for three types of verbs and negation
    elif (groupHello5 != None):
        output = "hi"
        if (groupHello5.group(1).lower() == "no quiero"):
            output = "Porque no quieres " + groupHello5.group(2) + "?"
        elif(groupHello5.group(1).lower() == "quiero"):
            output = "Porque quieres " + groupHello5.group(2) + "?"
        if (groupHello5.group(1).lower() == "no debo"):
            output = "Porque no debes " + groupHello5.group(2) + "?"
        elif(groupHello5.group(1).lower() == "debo"):
            output = "Porque debes " + groupHello5.group(2) + "?"
        if (groupHello5.group(1).lower() == "no puedo"):
            output = "Porque no puedes " + groupHello5.group(2) + "?"
        elif(groupHello5.group(1).lower() == "puedo"):
            output = "Porque puedes " + groupHello5.group(2) + "?"
    #checking for pienso or espero
    elif (groupHello6 != None):
        if (groupHello6.group(1).lower() == "pienso"):
            output = "Porqué piensas eso?"
        else:
            output = "Porqué esperas eso?"
    #checking for siempre
    elif (groupHello7 != None):
        output = "Puedes darme un ejemplo específico?"
    #checking for insult
    elif (groupHello8 != None):
        output = "Hey, sin insultos! Cálmate y cuéntame más."
    else:
        output = "Cuéntame más."
    return output

# =====================================================================================
# There are many ways that come to mind to improve Eliza. I've decided to pull 2 improvement from chapter 2 and one from chapter 21.
# Firstly, from section 2.6 on lemmatization and tokenization, we could help eliza with Contextual Understanding with Tokenization and Lemmatization
# We can incorporate lemmatization and context-aware tokenization to understand different forms of words
# and extract user intent more effectively. This will allow the program to understand variations in user input ( for e.g., "I am sad" vs. "I feel sadness")

# Second, we could introduce weighted pattern matching with edit distance.
# Section 2.8 talks about how we can use string similarity and use edit distance algorithms to make the program more tolerant of typos or
#  slight variations in input, ensuring more natural interactions (e.g., recognizing "happpy" as "happy").

# Finally, I really liked the idea of potentially using FrameNet to give eliza background knowledge about conversation topics.
# We can get eliza to recognize deeper relationships between words and respond based on the situation (e.g., understanding "buy" and "purchase" in the same context).
# I read more about FrameNet in section 21.5

# =====================================================================================

```

```python
# Request input from user and then print output
while True:
  userString = input("USER:     ")
  if userString == "quit":
    break
  output = eliza(userString)
  print("COMPUTER: " + output)
```
